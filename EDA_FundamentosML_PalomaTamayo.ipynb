

## Estadísticas generales con describe()

## Media por Fila y por Columna

```python
# Media por fila (de columnas numéricas)
df.select_dtypes(include=['int64', 'float64']).mean(axis=1).head()
```

## Tabla cruzada entre Rotación y Departamento

```python
pd.crosstab(df['Attrition'], df['Department'])
#Muestra a las personas que se quedaron o se fueron por departamento.
```

## Identificación de Variables

```python
cat_vars = df.select_dtypes(include='object').columns.tolist()
num_vars = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Variables categóricas:", cat_vars)
print("Variables numéricas:", num_vars)
#crea dos listas: una con variables categóricas y otra con numéricas.
```

## Medidas de Tendencia Central y Dispersión

```python
# Media, Mediana, Moda, Desviación estándar, Varianza, Cuartiles
print("Edad: Media =", df['Age'].mean())
print("Edad: Mediana =", df['Age'].median())
print("Edad: Moda =", df['Age'].mode()[0])
print("Edad: Desviación estándar =", df['Age'].std())
print("Edad: Varianza =", df['Age'].var())
print("Edad: Cuartiles =\n", df['Age'].quantile([0.25, 0.5, 0.75]))
```

## Histogramas de Variables Numéricas

```python
def plot_hist(variable):
    plt.figure(figsize=(9, 3))
    plt.hist(df[variable], bins=50)
    plt.xlabel(variable)
    plt.ylabel("Frecuencia")
    plt.title(f"Distribución de {variable}")
    plt.show()

numericVar = ["Age", "MonthlyIncome", "YearsAtCompany"]
for var in numericVar:
    plot_hist(var)
    #Crea histogramas personalizados para cada variable
```

## Diagramas de Caja (Boxplots)

```python
for var in numericVar:
    sns.boxplot(y=df[var])
    plt.title(f'Boxplot de {var}')
    plt.show()
    #muestra valores extremos y dispersión
```

## Frecuencia de Variables Categóricas

```python
for var in cat_vars:
    sns.countplot(data=df, x=var)
    plt.title(f'Distribución de {var}')
    plt.xticks(rotation=45)
    plt.show()
    #cuenta cuantos valores hay en cada categoria
```

## Análisis de Rotación de Personal

```python
# Crear columna binaria
df['Attrition_bin'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Gráfico de barras por JobLevel
g = sns.catplot(x="JobLevel", y="Attrition_bin", kind="bar", data=df, height=6)
g.set_ylabels("Probabilidad de Rotación")
plt.title("Rotación según Nivel de Cargo")
plt.show()
#convierte texto en número
```

## Boxplot de Ingreso por Rol

```python
df.boxplot(column="MonthlyIncome", by="JobRole")
plt.title("Ingreso mensual por Rol de Trabajo")
plt.xticks(rotation=45)
plt.ylabel("Ingreso Mensual")
plt.show()
#compara sueldos con roles de trabajo
```

## Relación entre Variables Numéricas

```python
sns.pairplot(df[["Age", "MonthlyIncome", "YearsAtCompany", "JobSatisfaction", "Attrition"]], hue="Attrition")
plt.show()
#Muestra correlaciones entre variables y si hay patrones en personas que rotan.
```

```python
from scipy import stats

grupos = [df[df['JobRole'] == rol]['MonthlyIncome'] for rol in df['JobRole'].unique()]
f_stat, p_val = stats.f_oneway(*grupos)

print("Estadístico F:", round(f_stat, 2))
print("Valor p:", round(p_val, 4))

if p_val < 0.05:
    print("✔️ Hay diferencias significativas entre los grupos.")
else:
    print("❌ No hay diferencias significativas.")
    # Detecta si el ingreso varía de forma significativa entre distintos roles.
```

```python
# Probabilidad de que alguien rote dado que está en ventas
total_ventas = df[df['Department'] == 'Sales'].shape[0]
rotan_en_ventas = df[(df['Department'] == 'Sales') & (df['Attrition'] == 'Yes')].shape[0]
prob = rotan_en_ventas / total_ventas
print(f"Probabilidad de rotación en ventas: {round(prob, 3)}")
```

## Conclusiones

- Empleados con menos antigüedad y menor ingreso tienen mayor probabilidad de rotar.
- Existen valores atípicos en `MonthlyIncome`.
- Las visualizaciones reflejan posibles asociaciones entre ingreso, antigüedad y rotación.
